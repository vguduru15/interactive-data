### Data Engineer – Hexbin Plot
A hexbin plot is a type of scatter plot where data points are aggregated into hexagonal bins. The color intensity of each bin represents the density of data points within it. A data engineer could apply this to analyze large datasets, such as tracking server load and network traffic, with hexagons showing areas where the highest volumes of data are concentrated.

#### Point 4
Generate an empty figure with two axes arranged in a matrix format, following the methods and syntax taught in class. Labelling the first axis as "Age Histogram" and the second axis as "Political Identity."

#### Point 5
Construct a figure featuring a single axis that presents a bar plot illustrating the frequency of responses for the Evaluation variable.\

```python
# Summarize and plot the frequencies of the Evaluation variable
evaluation_counts = human_data['Evaluation'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
evaluation_counts.plot(kind='bar', ax=ax)
ax.set_title('Frequencies of Evaluation Responses')
ax.set_xlabel('Evaluation')
ax.set_ylabel('Frequency')

plt.show()
```

#### Point 6
Construct a figure featuring a single axis that shows a scatter plot representing the ConMediaUse and ImmiPhobia variables.

```python
# Scatter plot for ConMediaUse and ImmiPhobia
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(human_data['ConMediaUse'], human_data['ImmiPhobia'])
ax.set_title('ConMediaUse vs. ImmiPhobia')
ax.set_xlabel('ConMediaUse')
ax.set_ylabel('ImmiPhobia')

plt.show()
```
