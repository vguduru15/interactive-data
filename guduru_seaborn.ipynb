```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv("titanic_clean.csv")
```

```python
titanic.head(10)
```

```python
titanic.head()
```

```python
sns.relplot(
        data=titanic,
        x="age", y="fare", col="survived",
        hue="sex", size="pclass",
        palette={"male": "blue", "female": "pink"},
)
```

```python
sns.relplot(
        data=titanic,
        x="age", y="fare", col="survived",
        hue="sex", size="pclass",
        palette={"male": "blue", "female": "pink"},
        alpha =0.7
)
```

```python
sns.relplot(
    data=titanic,
    x="age", y="fare", col="sex",
    hue="survived", size="pclass",
)
```

```python
titanic['survive_ind'] = titanic['survived'].map({0: "didn't survive", 1:"survived"})



sns.relplot(
        data=titanic,
        x="age", y="fare", col="sex",
        hue="survive_ind", size="pclass",
        palette={"didn't survive": "red", "survived": "green"},
        alpha =0.7
)
```

```python
sns.relplot(
    data = titanic,
    x = "sex", y = "alone", hue = "embark_town"
)
```

```python
titanic.groupby('embark_town')['alone'].describe()
```

```python
sns.relplot(
    data = titanic,
    x = "embark_town", y = "sex", hue ="alone"
)
```

```python
smh.groupby("Choose your gender")[['Do you have Depression?','Do you have Anxiety?','Do you have Panic attack?']].count()
```

```python
columns_to_plot = smh.columns[-4:-1]

category_order = ['Yes', 'No']
```

```python
for column in columns_to_plot:
    counts = smh[column].value_counts()

    counts = counts.reindex(category_order, fill_value=0)


print(counts)
print()
```
